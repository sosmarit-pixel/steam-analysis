```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the CSV file
df = pd.read_csv('reviews_BlackOps7.csv')

# Count words in each review
df['word_count'] = df['review'].apply(lambda x: len(str(x).split()))

# Create histogram with log scale x-axis
plt.figure(figsize=(12, 6))
plt.hist(df['word_count'], bins=50, edgecolor='black', alpha=0.7)
#plt.yscale('log')
plt.xlabel('Number of Words', fontsize=12)
plt.ylabel('Number of Reviews', fontsize=12)
plt.title('Distribution of Review Lengths', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3, which='both')
plt.tight_layout()
plt.show()

# Display some statistics
print(f"Total reviews: {len(df)}")
print(f"\nWord count statistics:")
print(df['word_count'].describe())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the CSV file
df = pd.read_csv('reviews_BlackOps7.csv')

# Count words in each review
df['word_count'] = df['review'].apply(lambda x: len(str(x).split()))

# Filter to only reviews with < 200 words
df_filtered = df[df['word_count'] < 200]

# Create histogram with filtered data
plt.figure(figsize=(12, 6))
plt.hist(df_filtered['word_count'], bins=100, edgecolor='black', alpha=0.7)
plt.xlabel('Number of Words', fontsize=12)
plt.ylabel('Number of Reviews', fontsize=12)
plt.title(f'Distribution of Review Lengths (< 200 words, n={len(df_filtered)})', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3, which='both')
plt.tight_layout()
plt.show()

print(f"Total reviews: {len(df)}")
print(f"Reviews with < 200 words: {len(df_filtered)}")
```
